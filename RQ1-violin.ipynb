```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ==========================================
# 1. SETUP AND DATA LOADING
# ==========================================
df = pd.read_csv("FoAI_A2_data_4.6k.csv")
sns.set(style="whitegrid")

# ==========================================
# 2. CUSTOM LEGEND MAPPING (Your Request)
# ==========================================
# Map the codes to the full names you want in the legend
experience_map = {
    'EN': 'Entry Level',
    'MI': 'Mid Level',
    'SE': 'Senior Level',
    'EX': 'Executive Level'
}

# Apply the mapping to create a new column for the legend
df['Experience Label'] = df['experience_level'].map(experience_map)

# Define the logical order for the legend using the NEW names
experience_order = ['Entry Level', 'Mid Level', 'Senior Level', 'Executive Level']

# Map Remote Ratio labels for clarity
df['Work Arrangement'] = df['remote_ratio'].map({0: 'On-site', 50: 'Hybrid', 100: 'Remote'})

# ==========================================
# 3. VIOLIN PLOT GENERATION
# ==========================================
plt.figure(figsize=(14, 8))

sns.violinplot(
    x='Work Arrangement',
    y='salary_in_usd',
    hue='Experience Label',  # Use the new column with full names
    data=df,
    hue_order=experience_order, # Ensure the order is correct (Entry -> Executive)
    palette='viridis',
    split=False,
    inner='quartile'
)

# Titles and formatting
plt.title('Salary Distribution by Remote Type and Experience Level', fontsize=16)
plt.xlabel('Work Arrangement', fontsize=12)
plt.ylabel('Salary (USD)', fontsize=12)
plt.legend(title='Experience Level', loc='upper left')

plt.show()
```
